```python
import requests  # Para fazer requisições HTTP
import pandas as pd  # Para manipulação de dados
import yfinance as yf  # Para obter dados financeiros
```

```python
#Lista de tickers dos ativos
tickers = pd.read_csv("tickers.csv")
#Converte para uma lista com base na coluna Código
tickers = tickers['Código'].to_list()
start_date = '2017-01-01'
end_date = '2023-11-19'
#Pega a coluna do Preço Ajustado, porque pegar preço considerando dividentos e outros fatores
data = yf.download(tickers, start=start_date, end=end_date)['Close']
#a coluna que tem a data é a index, criando no final a coluna time
data['time'] = data.index
data
```

```python
#Use a função melt para transformar o DataFrame
data = pd.melt(data, id_vars=['time'], var_name='Ativo', value_name='Preço')
data
```

```python
#calcula o retono dos 12 meses
#Agrupa Preços do Ativo calculando a variação dentro de 12 meses.
data['retorno_12m'] = data.groupby('Ativo')['Preço'].transform(lambda x: x / x.shift(252) - 1)
#cria uma coluna com o valor do preço do dia anterior
data['Lagged_Preco'] = data.groupby('Ativo')['Preço'].shift()
data
```

```python
#Filtra por ativo a linha do último dia de cada mês
#Converte a coluna 'time' para o tipo de datatime se ainda não for
data['time'] = pd.to_datetime(data['time'])

#Use a função groupby para agrupar por 'Ativo' e o ano_mês do 'time'
groups = data.groupby(['Ativo', data['time'].dt.to_period('M')])

#use a função idmax para obter o índice do máximo para cada grupo
idx = groups['time'].idxmax()

#Crie um DataFrame 'data_mensal' usando os índices encontrados
data_mensal = data.loc[idx]
data_mensal
```

```python
#Criando coluna ano_mes
data_mensal['Yearmon'] = data_mensal['time'].dt.to_period('M')
data_mensal
```
